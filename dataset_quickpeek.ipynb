```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from taskdataset import TaskDataset

path_to_data = "datasets/" # TODO Set path to datasets files
```

```python
dataset = torch.load(path_to_data+"ModelStealingPub.pt")
print(f"DATASET LEN: [{len(dataset.labels)}] UNIQUE LABELS: [{len(np.unique(dataset.labels))}]")
lim = 5
cnt = 0
for id, im, lb in zip(dataset.ids, dataset.imgs, dataset.labels):
    plt.imshow(im)
    plt.title(f"ID [{id}] LABEL [{lb}] SIZE {np.array(im).shape}")
    plt.axis("off")
    plt.show()
    cnt += 1
    if cnt >= lim: break
```
